```python
# Import necessary libraries.
import folium
import json
import pandas
```

### Part 1: Getting Some Data

```python
# Fetch the data from the source link.
road_closures = pandas.read_csv("https://opendata.maryland.gov/resource/nigh-m2sg.csv")

# This code block fetches and displays the dataset from opendata maryland that consists of data relating to road closures in the state of Maryland. This data includes the county of the road closure, the road of the incident, the direction, the number of lanes closed down and which direction those lanes are headed, the exact location, and the data and time the item was added to and updated in the dataset.
```

### Part 2: Making a Map

```python
# Create a map and center it on Washington D.C so the state of Maryland shows fully.
map_osm = folium.Map(location=[38.9, -77.04], zoom_start=9)
```

### Part 3: Combining Parts 1 and 2

```python
# Visit every row of the road closure data set.
for index, closure in road_closures.iterrows():
    # Collect the latitude and longitude from the row.
    coords = json.loads(closure["long"])

    # Plot all northbound road closures on the map in their respective color and location.
    if closure["direction"] == "North":
        folium.Marker(location=[coords["y"], coords["x"]],
                    icon=folium.Icon(color='red')).add_to(map_osm)
    # Plot all southbound road closures on the map in their respective color and location.    
    if closure["direction"] == 'South':
        folium.Marker(location=[coords["y"], coords["x"]],
            icon=folium.Icon(color='blue')).add_to(map_osm)
    # Plot all westbound road closures on the map in their respective color and location.
    if closure["direction"] == 'West':
        folium.Marker(location=[coords["y"], coords["x"]],
            icon=folium.Icon(color='black')).add_to(map_osm)
    # Plot all eastbound road closures on the map in their respective color and location.
    if closure["direction"] == 'East':
        folium.Marker(location=[coords["y"], coords["x"]],
            icon=folium.Icon(color='green')).add_to(map_osm)

# Show the map with all of it's plots.
map_osm 

# This code seperates out and plots all road closures based on the direction of the closed road. This map helps to display the areas in MD that have closed road and can help depict which direction on said road there will be traffic. It can also help to depict which sides of certain roads see the most traffic, as some places may show a road with accidents only one direction. For example, If a road has accidents in the eastbound direction, but not the westbound direction, one can come to certain conclusions about the traffic on that road.
```
